```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score

# URLs for the datasets
country_code_url = "https://github.com/FlipRoboTechnologies/ML_-Datasets/raw/main/Z_Restaurant/Country-Code.xlsx"
zomato_url = "https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Z_Restaurant/zomato.csv"

# Load country code data
country_code_df = pd.read_excel(country_code_url)

# Load Zomato restaurant data with specific encoding
zomato_df = pd.read_csv(zomato_url, encoding='latin1')

# Check the column names to verify exact spelling and format
print(zomato_df.columns)

# Ensure there are no leading/trailing spaces in column names
zomato_df.columns = zomato_df.columns.str.strip()

# Example: Perform EDA and Feature Engineering
# Example: Pairplot for initial exploration
sns.pairplot(zomato_df[['Aggregate rating', 'Votes', 'Average Cost for two']], diag_kind='kde')
plt.suptitle('Pairplot for EDA', y=1.02)
plt.show()

# Example: Correlation Heatmap
# Exclude non-numeric columns like 'Cuisines' from correlation heatmap
numeric_columns = ['Aggregate rating', 'Votes', 'Average Cost for two']
plt.figure(figsize=(10, 8))
sns.heatmap(zomato_df[numeric_columns].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# Example: Feature Selection
X = zomato_df[['Aggregate rating', 'Votes']]  # Features for prediction
y_cost = zomato_df['Average Cost for two']  # Target variable for cost prediction
y_price_range = zomato_df['Price range']  # Target variable for price range prediction

# Split data into training and testing sets
X_train, X_test, y_train_cost, y_test_cost = train_test_split(X, y_cost, test_size=0.2, random_state=42)
X_train, X_test, y_train_price_range, y_test_price_range = train_test_split(X, y_price_range, test_size=0.2, random_state=42)

# Example: Model Training
cost_model = LinearRegression()
cost_model.fit(X_train, y_train_cost)

price_range_model = LogisticRegression(max_iter=1000)
price_range_model.fit(X_train, y_train_price_range)

# Example: Model Evaluation
cost_predictions = cost_model.predict(X_test)
price_range_predictions = price_range_model.predict(X_test)

cost_rmse = mean_squared_error(y_test_cost, cost_predictions, squared=False)
price_accuracy = accuracy_score(y_test_price_range, price_range_predictions)

print(f"Root Mean Squared Error (Cost prediction): {cost_rmse}")
print(f"Accuracy (Price range prediction): {price_accuracy}")
```

```python
# Example: Model Evaluation and Fine-tuning

# Example: Model Evaluation
# Already computed RMSE and accuracy in previous steps

print(f"Root Mean Squared Error (Cost prediction): {cost_rmse}")
print(f"Accuracy (Price range prediction): {price_accuracy}")

# Example: Fine-tuning (Hyperparameter optimization)

from sklearn.model_selection import GridSearchCV

# Example: Fine-tuning Logistic Regression for price range prediction
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularization parameter
    'penalty': ['l1', 'l2']  # Regularization type
}

grid_search = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)
grid_search.fit(X_train, y_train_price_range)

print("Best parameters found for Logistic Regression:")
print(grid_search.best_params_)

best_price_range_model = grid_search.best_estimator_

# Example: Re-evaluate with best model
best_price_range_predictions = best_price_range_model.predict(X_test)
best_price_accuracy = accuracy_score(y_test_price_range, best_price_range_predictions)
print(f"Accuracy with best model (Price range prediction): {best_price_accuracy}")

# Example: Interpretation and Insights

# Example: Feature Importance (if applicable)
# Logistic Regression can provide coefficients as feature importance
if isinstance(best_price_range_model, LogisticRegression):
    feature_importance = pd.Series(best_price_range_model.coef_[0], index=X.columns)
    feature_importance.plot(kind='barh')
    plt.title('Feature Importance for Price Range Prediction')
    plt.xlabel('Coefficient Magnitude')
    plt.show()

# Example: Deployment or Reporting
# Deploy the model if applicable, or prepare a detailed report summarizing findings and insights.

```
